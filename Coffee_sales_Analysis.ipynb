```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Coffe_sales.csv")
df.head(10)
```

```python
df.drop_duplicates(inplace=True)
df.head(2)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Time"] = df["Time"].str.split(".").str[0]
df.head(3)
```

```python
df.columns = df.columns.str.capitalize().str.replace(" ","_")
df.head(3)
```

```python
df = pd.get_dummies(df, columns=["Time_of_day"], drop_first=True)
```

```python
print(df[["Hour_of_day", "Money"]].describe())
```

```python
df.to_csv("Coffee_sales_cleaned.csv", index=False)
print("Cleaned dataset saved")
df.head(16)
```
